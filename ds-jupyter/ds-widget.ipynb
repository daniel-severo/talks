```python
import ipywidgets
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-1, 1, 1000)
```

```python
def show_sin(f):
    y = np.sin(2*np.pi*f*x)
    plt.plot(x,y)
    plt.show()
```

```python
ipywidgets.interact(show_sin, f=(1.0,10.0))
```
