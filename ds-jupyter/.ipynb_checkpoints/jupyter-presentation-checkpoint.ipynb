# <center>Easy as py.</center>

```python
# imports.
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# define variables.
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

# plots!
plt.plot(x,y)
plt.show()
```

# <center>Podemos fazer melhor?</center>

```python
# add some flavor.
π = np.pi #\pi
```

```python
# redefine
x = np.linspace(0, 4*π, 100)
```

```python
# plot again
plt.style.use('ggplot')
plt.figure(figsize=(16,6))
plt.plot(x,y)
plt.show()
```
